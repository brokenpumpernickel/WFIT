```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
def func(y, t, w0):
    x, v = y
    return np.array([v, -w0**2*x])
```

```python
def func(y, t, w0):
    a = np.empty(2)
    a[0] = y[1]
    a[1] = -w0**2*y[0]
    return a
```

```python
x0, v0 = 1, 0
w0 = 1
ts = np.linspace(0, 20, 1000)
res = odeint(func, (x0, v0), ts, args = (w0,))
plt.plot(res[:, 0], res[:, 1])
res
```

```python
plt.figure(figsize=(20, 10))
plt.xlim(0, 200)
plt.plot(ts, res[:, 0])
```

```python
ts = np.linspace(0, 20, 100000)
x0, v0 = 1, 0
w0 = 1
res = np.zeros((len(ts), 2))
res[0, :] = x0, v0
for w in range(1, len(ts)):
    res[w, :] = res[w - 1, :] + func(res[w - 1, :], ts[w - 1], w0) * (ts[w] - ts[w - 1])
```

```python
plt.plot(ts, res[:, 0])
```

```python
plt.plot(res[:, 0], res[:, 1])
```

```python
func(1, 1, 1)
```

```python
quad(func, -1, 1, (1, 1))
```
